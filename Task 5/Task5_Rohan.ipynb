```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.findeasy.in/tobacco-use-among-adults-in-india/"
response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, "html.parser")

# Find all tables on the webpage
tables = soup.find_all("table")

# Check if the second table exists
if len(tables) > 1:
    first_table = tables[0]
    
    states = []
    women = []
    men = []

    for row in first_table.find_all("tr")[1:]:
        cells = row.find_all("td")
        state = cells[1].text.strip()
        women_age = str(cells[2].text.strip().replace(",", ""))
        men_age = str(cells[3].text.strip().replace(",", ""))
        states.append(state)
        women.append(women_age)
        men.append(men_age)

    df = {
        "State": states,
        "Women_age_15_and_above": women,
        "Men_age_15_and_above": men,
    }
    data = pd.DataFrame(df)
    print("First Table:")
    print(data)
else:
    print("The first table was not found on the webpage.")
```

```python
data.drop(36,axis = 0, inplace = True)
```

```python
new = data["Women_age_15_and_above"].str.split("%", n = 1, expand = True)
data["Women_age_15_and_above"]= new[0]
new = data["Men_age_15_and_above"].str.split("%", n = 1, expand = True)
data["Men_age_15_and_above"]= new[0] 
```

```python
data['Women_age_15_and_above'] = data['Women_age_15_and_above'].astype(float)
data['Men_age_15_and_above'] = data['Men_age_15_and_above'].astype(float)
```
